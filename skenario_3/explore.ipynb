```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
import joblib
from sklearn.preprocessing import MinMaxScaler
```

```python
RESULT_FINAL_DIR = 'result/'
MODEL_FINAL_DIR = 'models/'
DATA_FINAL_DIR = 'data/'
target = 'islulus'
```

```python
train_path = os.path.join(
    DATA_FINAL_DIR, 'train.csv')
test_path = os.path.join(DATA_FINAL_DIR, 'test.csv')

train = pd.read_csv(train_path)
train.replace([np.inf, -np.inf], np.nan, inplace=True)
train.dropna(inplace=True)
test = pd.read_csv(test_path)

X = train.drop(columns=[target])
y = train[target]
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
X_test = scaler.transform(test.drop(columns=[target]))
y_test = test[target]

joblib.dump(scaler, os.path.join(MODEL_FINAL_DIR, 'scaler.joblib'))
```

```python
result_cnn = pd.read_csv(os.path.join(RESULT_FINAL_DIR, 'results_cnn.csv'))
result_dense = pd.read_csv(os.path.join(RESULT_FINAL_DIR, 'results_dense.csv'))
```

```python
result_cnn_mean = result_cnn.groupby(['model', 'optimizer', 'learning_rate', 'batch_size']).mean().reset_index()
result_dense_mean = result_dense.groupby(['model', 'optimizer', 'learning_rate', 'batch_size']).mean().reset_index()
```

```python
history = pd.read_csv(os.path.join(MODEL_FINAL_DIR, 'history_cnn_rmsprop_0.0001_batch16_fold3.csv'))
```

```python
history.reset_index(inplace=True)
history['epoch'] = history['index'] + 1
```

```python
# visualize history

plt.figure(figsize=(10, 6))
plt.plot(history['epoch'], history['accuracy'], label='Accuracy')
plt.plot(history['epoch'], history['val_accuracy'], label='Validation Accuracy')
plt.title('CNN Model Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

```python
# visualize loss
plt.figure(figsize=(10, 6))
plt.plot(history['epoch'], history['loss'], label='Loss')
plt.plot(history['epoch'], history['val_loss'], label='Validation Loss')
plt.title('CNN Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
```
