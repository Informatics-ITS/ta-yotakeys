```python
import pandas as pd
import os
```

```python
DATA_FINAL_DIR = 'data/'
MODEL_FINAL_DIR = 'models/'
RESULT_FINAL_DIR = 'result/'
```

```python
list_model = [
    'RandomForest', 'XGBoost', 'SVM', 'LogisticRegression', 'AdaBoost', 'ExtraTrees', 'DecisionTree'
]


result = pd.DataFrame()

for model in list_model:

    result_file = os.path.join(RESULT_FINAL_DIR, f'results_{model}.csv')

    if os.path.exists(result_file):

        df = pd.read_csv(result_file)

        df['model'] = model

        result = pd.concat([result, df], ignore_index=True)
```

```python
result
```

```python
result.groupby(['model']).mean().reset_index()
```

```python
result.to_csv(os.path.join(RESULT_FINAL_DIR, 'result.csv'), index=False)
```

```python
result.to_excel(os.path.join(RESULT_FINAL_DIR, 'result.xlsx'), index=False)
```

```python
result.groupby(['model']).mean().reset_index().sort_values(by=['f1_score'], ascending=False)
```

```python
best_result = result.groupby(['model']).mean().reset_index()
```
