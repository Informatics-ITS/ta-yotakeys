```python
result = result.groupby('skenario').mean().reset_index()
```

```python
# map skenario to 3  new columns

def new_columns(row):
    if row['skenario'] == '_':
        return pd.Series(['Not Clean', 'No', 'Ordinal'])
    elif row['skenario'] == '_clean_':
        return pd.Series(['Clean', 'No', 'Ordinal'])
    elif row['skenario'] == '_clean_feature_':
        return pd.Series(['Clean', 'Yes', 'Ordinal'])
    elif row['skenario'] == '_clean_feature_onehot_1_':
        return pd.Series(['Clean', 'Yes', 'Mix'])
    elif row['skenario'] == '_clean_feature_onehot_2_':
        return pd.Series(['Clean', 'Yes', 'One Hot'])
    elif row['skenario'] == '_clean_onehot_1_':
        return pd.Series(['Clean', 'No', 'Mix'])
    elif row['skenario'] == '_clean_onehot_2_':
        return pd.Series(['Clean', 'No', 'One Hot'])
    elif row['skenario'] == '_feature_':
        return pd.Series(['Not Clean', 'Yes', 'Ordinal'])
    elif row['skenario'] == '_feature_onehot_1_':
        return pd.Series(['Not Clean', 'Yes', 'Mix'])
    elif row['skenario'] == '_feature_onehot_2_':
        return pd.Series(['Not Clean', 'Yes', 'One Hot'])
    elif row['skenario'] == '_onehot_1_':
        return pd.Series(['Not Clean', 'No', 'Mix'])
    elif row['skenario'] == '_onehot_2_':
        return pd.Series(['Not Clean', 'No', 'One Hot'])
    
result[['clean', 'feature', 'encoding']] = result.apply(new_columns, axis=1)
        
```

```python
result = pd.read_csv(os.path.join(RESULT_FINAL_DIR, 'result.csv'))
```

```python
result[['clean', 'feature', 'encoding']] = result.apply(new_columns, axis=1)
```

```python

# get best result for each clean, feature, encoding combination
# drop skenario and fold columns
# change clean, feature, encoding, to the first columns
# make 4

best_results = result.loc[result.groupby(['clean', 'feature', 'encoding'])['f1_score'].idxmax()].reset_index(drop=True)
best_results.drop(columns=['skenario', 'fold'], inplace=True)
best_results = best_results[['clean', 'feature', 'encoding', 'accuracy', 'precision', 'recall', 'f1_score', 'confusion_matrix']]
```

```python
best_results
```

```python
best_results.drop(columns='confusion_matrix').to_excel(os.path.join(RESULT_FINAL_DIR, 'best_results.xlsx'), index=False)
```

```python
import numpy as np
import ast
import matplotlib.pyplot as plt
import seaborn as sns

best_rows = best_results

# Buat grid 6 baris x 2 kolom (12 subplot)
fig, axs = plt.subplots(6, 2, figsize=(12, 24))  # Ukuran fig dibuat vertikal
axs = axs.flatten()

for i, (index, row) in enumerate(best_rows.iterrows()):
    if i >= 12:
        break  # maksimal 12 model

    cm = np.array(ast.literal_eval(row['confusion_matrix']))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        cbar=False,
        xticklabels=['Predicted Negative', 'Predicted Positive'],
        yticklabels=['Actual Negative', 'Actual Positive'],
        ax=axs[i],
        annot_kws={"size": 14}
    )

    # Judul digabung dari 3 kolom
    title = f"clean={row['clean']} | feature={row['feature']} | encoding={row['encoding']}"
    axs[i].set_title(title, fontsize=14)
    axs[i].set_xlabel('Predicted', fontsize=12)
    axs[i].set_ylabel('Actual', fontsize=12)
    axs[i].tick_params(axis='both', labelsize=10)

# Hapus subplot kosong jika model < 12
for j in range(i + 1, 12):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.suptitle("Confusion Matrices by Preprocessing Configuration", fontsize=20, y=1.02)
plt.show()
```

```python
MODEL_FINAL_DIR = 'models/'
result = pd.read_csv(os.path.join(RESULT_FINAL_DIR, 'result.csv'))
result[['clean', 'feature', 'encoding']] = result.apply(new_columns, axis=1)
best_rows = result.loc[result.groupby(['clean', 'feature', 'encoding'])[
    'f1_score'].idxmax()].reset_index(drop=True)
```

```python
best_rows
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import os

fig, axs = plt.subplots(6, 2, figsize=(14, 12))
axs = axs.flatten()  # Flatten for easy indexing

for i, (index, row) in enumerate(best_rows.iterrows()):
    if i >= 12:
        break 

    FILE_HISTORY = os.path.join(
        MODEL_FINAL_DIR,
        f"history_{row['skenario']}_fold{row['fold']}.csv"
    )

    history = pd.read_csv(FILE_HISTORY)
    history['epoch'] = history.index + 1

    axs[i].plot(history['epoch'], history['loss'], label='Loss', color='blue')
    axs[i].plot(history['epoch'], history['val_loss'],
                label='Validation Loss', color='orange')
    axs[i].set_title(
        f"{row['clean']} | {row['feature']} | {row['encoding']}",)
    axs[i].set_xlabel('Epoch')
    axs[i].set_ylabel('Loss')
    axs[i].legend()
    axs[i].grid()

for j in range(i+1, 12):
    fig.delaxes(axs[j])

plt.tight_layout()
plt.suptitle("Loss History for Best Models", fontsize=16, y=1.02)
plt.show()
```
